### Activation Function

```python
# sigmoid = 1/1+e^-x

def sigmoid(x):
   return 1/(1+np.exp(-x))
```

### Inializing the weights and bias

```python
def initialize(parameters1):

    w = np.zeros((parameters1,1))
    b = 0.0
    

    return w,b
```

### Forward and Backward Propagation

### Forward Propagation    --> 𝐴 = 𝜎(𝑤𝑇𝑋+𝑏)=(𝑎(1),𝑎(2),...,𝑎(𝑚−1),𝑎(𝑚))
### Cost Function          --> 𝐽 = −1/𝑚 * ∑(𝑦(𝑖)log(𝑎(𝑖))+(1−𝑦(𝑖))log(1−𝑎(𝑖)))
### Backward Propagation   --> ∂𝐽/∂𝑤 = 1/𝑚 * 𝑋(𝐴−𝑌)𝑇 and ∂𝐽/∂𝑏 = 1/𝑚 * ∑(𝑎(𝑖)−𝑦(𝑖))

```python
def propagation(w,b,X,Y):
    
    #FORWARD PROPAGATION
    #--------------------#

    m = X.shape[1]

    Z = np.dot(w.T,X)+b
    a =  sigmoid(Z)
   
    cost = -(1/m)*np.sum(Y*np.log(a)+(1-Y)*np.log(1-a))

    #BACKWARD PROPAGATION
    #--------------------#

    dw = (1/m) * np.dot(a-Y,X.T)
    db = (1/m) * np.sum(a-Y)

    #cost = np.squeeze(np.array(cost))
    #--------------------------------#

    grads = {
        "dw" : dw,
        "db" : db
    }
    return grads,cost
```

### Gradient Descent --> 𝜃=𝜃−𝛼 𝑑𝜃
 

```python
def gradient_descent(w,b,X,Y,num_iterations = 1000,learning_rate = 0.009):
    
    costs = []

    for i in range(num_iterations):
        grads,cost = propagation(w, b, X,Y)

        dw = grads['dw']
        db = grads['db']

        w = w - learning_rate*dw.T
        b = b - learning_rate*db
        if(i>20 and i <5470):
            continue
        print("Cost after iteration ",i ,"is : ",cost)
        costs.append(cost)

    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}
    return params,grads,cost
```

### Predicton -->  0 (if activation <= 0.5) or 1 (if activation > 0.5)

```python
def predict(w,b,X):

    z= np.dot(w.T,X)+b
    predictions = sigmoid(z)

    predictions = predictions >=0.5
    predictions = np.array(predictions, dtype='int64')
    return predictions
```

### --------LOGISTIC REGRESSION MODEL--------

```python
def logisticRegression(X_train,Y_train,X_test,Y_test, num_iterations = 10000,learning_rate = 0.05):

    w,b = initialize(X_train.shape[0])
    parms,grads,costs = gradient_descent(w,b,X_train,Y_train,num_iterations = num_iterations,learning_rate = learning_rate)
    w = parms['w']
    b = parms['b']
    
    predictions_train = predict(w,b,X_train)
    predictions_test = predict(w,b,X_test)

    print("\n\n\nThe training accuracy of the model is " , 100 - np.mean(np.abs(predictions_train-Y_train))*100)
    print("The training accuracy of the model is ",100- np.mean(np.abs(predictions_test-Y_test))*100)
```
